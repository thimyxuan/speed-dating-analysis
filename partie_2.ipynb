```python
wave_list, male_list, female_list  = [], [], []
for i in range(1,22):
    wave_list.append(i)
    male_list.append(len(data_initial.iid[data_initial.wave == i][data_initial.gender == 0].unique()))
    female_list.append(len(data_initial.iid[data_initial.wave == i][data_initial.gender == 1].unique()))

wave_distribution = {'wave': wave_list, 'male': male_list, 'female': female_list}
wave_distribution_df = pd.DataFrame(wave_distribution)
```

```python
# Habitudes sociales
social_habbits_df = data.groupby('iid').max()['go_out'].reset_index()
social_habbits_count_df = social_habbits_df['go_out'].value_counts().to_frame().reset_index()
social_habbits_count_df['frequency_name'] = social_habbits_count_df['go_out'].apply( 
                            lambda x : "Several times a week" if x == 7.0
                            else "Twice a week" if x == 6.0
                            else "Once a week" if x == 5.0
                            else "Twice a month" if x == 4.0
                            else "Once a month" if x == 3.0
                            else "Several times a year" if x == 2.0
                            else "Almost never" if x == 1.0
                            else x
)
print('Most frequent value (go_out): \n', social_habbits_count_df.sort_values(by='count', ascending=False)['frequency_name'][0])
```

```python
# Habitudes de dating
dating_habbits_df = data.groupby('iid').max()['date'].reset_index()
dating_habbits_count_df = dating_habbits_df['date'].value_counts().to_frame().reset_index()
dating_habbits_count_df['frequency_name'] = dating_habbits_count_df['date'].apply( 
                            lambda x : "Several times a week" if x == 7.0
                            else "Twice a week" if x == 6.0
                            else "Once a week" if x == 5.0
                            else "Twice a month" if x == 4.0
                            else "Once a month" if x == 3.0
                            else "Several times a year" if x == 2.0
                            else "Almost never" if x == 1.0
                            else x
)
print('Most frequent value (date): \n', dating_habbits_count_df.sort_values(by='count', ascending=False)['frequency_name'][0]
)
```

```python
# Nous rassemblons les deux donn√©es pour les visualiser ensemble
dating_habbits_count_df.rename(columns={'count': 'dating_count'}, inplace = True)
social_habbits_count_df.rename(columns={'count': 'social_count'}, inplace = True)
habbits_df = pd.merge(social_habbits_count_df, dating_habbits_count_df, left_on='go_out', right_on='date')
habbits_df = habbits_df.sort_values(by='go_out', ascending=False)
to_drop = ['date','go_out','frequency_name_x']
habbits_df.drop(to_drop, axis=1, inplace=True)
habbits_df.rename(columns={'frequency_name_y': 'frequency_name'}, inplace=True)
```

## Satisfaction de l'exp√©rience

```python
# Satisfaction avec les personnes rencontr√©es
print('Mean value (satis_2):', round(data.groupby('iid').max()['satis_2'].median(),2))
sns.set(rc={"figure.figsize":(4.5, 3)}, style = 'white')
sns.histplot(data=data.groupby('iid').max(), x='satis_2', hue='match', kde=False, bins=9, palette=['lime', 'blue'], edgecolor='black')
plt.legend(labels=['match','no match'], fontsize=10)
plt.ylabel('count', fontsize=11)
plt.xlabel('rating on 1-10 scale', fontsize=11)
plt.xticks([i for i in range (1,11)], fontsize=10)
plt.title('Satisfaction with people met (satis_2)', fontsize=12.5, pad=16)
plt.show()
```

```python
# Satisfaction avec le format des speed dates (nombre et dur√©e)
wavesize_average = data.groupby('wave').max()['round'].median()

speeddate_nb = data[data['round'] <= wavesize_average].groupby('iid').max()['numdat_2'].value_counts().to_frame().reset_index()
speeddate_len = data[data['round'] <= wavesize_average].groupby('iid').max()['length'].value_counts().to_frame().reset_index()
speeddate2_nb = data[data['round'] > wavesize_average].groupby('iid').max()['numdat_2'].value_counts().to_frame().reset_index()
speeddate2_len = data[data['round'] > wavesize_average].groupby('iid').max()['length'].value_counts().to_frame().reset_index()

speeddate_df = pd.merge(speeddate_nb, speeddate_len, left_on='numdat_2', right_on='length')
speeddate_df.drop(['length'], axis=1, inplace=True)
speeddate_df.rename(columns={'numdat_2': 'val', 'count_x': 'numdat_2', 'count_y': 'length'}, inplace=True)
speeddate_df = speeddate_df.sort_values(by='val')

speeddate_df2 = pd.merge(speeddate2_nb, speeddate2_len, left_on='numdat_2', right_on='length')
speeddate_df2.drop(['length'], axis=1, inplace=True)
speeddate_df2.rename(columns={'numdat_2': 'val', 'count_x': 'numdat_2', 'count_y': 'length'}, inplace=True)
speeddate_df2 = speeddate_df2.sort_values(by='val')
```

```python
# Visualisation avec Seaborn et Matplotlib
print('Average wave size:', wavesize_average)
sns.set(rc={'figure.figsize':(12, 3)}, style='white')
fig, axes = plt.subplots(1, 2)
fig.subplots_adjust(hspace=0.125, wspace=0.125)
fig.suptitle('Satisfaction with speed dates format', fontsize=12.5)
fig.subplots_adjust(top=.75)

speeddate_df.set_index('val').plot(kind='bar', stacked=False, color=['darkorange', 'deeppink'], width=.7, edgecolor='black', ax = axes[0])
speeddate_df2.set_index('val').plot(kind='bar', stacked=False, color=['darkorange', 'deeppink'], width=.7, edgecolor='black', ax = axes[1])

axes[0].set_title('Number of people met <= 14', fontsize=10, fontweight='bold')
axes[1].set_title('Number of people met > 14', fontsize=10, fontweight='bold')

axes[0].set_xlabel('')
axes[1].set_xlabel('')
axes[0].set_ylabel('count', fontsize=11)
axes[0].set_ylabel('')

axes[0].set_xticks([i for i in range(0,3)], ['Too few | Too short', 'Too many | Too long', 'Right qty | Right length'], rotation=0, fontsize = 8, fontweight='bold')
axes[1].set_xticks([i for i in range(0,3)], ['Too few | Too short', 'Too many | Too long', 'Right qty | Right length'], rotation=0, fontsize = 8, fontweight='bold')

axes[0].legend(labels=['number','length'], fontsize=10)
axes[1].legend(labels=['number','length'], fontsize=10)

plt.show()
```
